# **Feature Engineering**


1. **What is a parameter?**
     - A parameter is a variable in a machine learning model that is learned from the training data, like weights in linear regression.

2. **What is correlation? What does negative correlation mean?**
     - Correlation is a statistical measure that describes the degree to which two variables move in relation to each other. Negative correlation means that as one variable increases, the other decreases.

3. **Define Machine Learning. What are the main components in Machine Learning?**
     - Machine Learning is a field of AI where systems learn from data to make predictions. Key components include data, algorithms, models, and evaluation metrics.



4. **How does loss value help in determining whether the model is good or not?**
     - Loss indicates how far predictions are from actual values. A lower loss value means better model performance.

5. **What are continuous and categorical variables?**
     - Continuous variables are numerical and can take any value (e.g., height), while categorical variables represent groups or categories (e.g., gender).



6. **How do we handle categorical variables in Machine Learning? What are the common techniques?**
     - Techniques include One-Hot Encoding, Label Encoding, and using embeddings to convert categories into numerical form.

7. **What do you mean by training and testing a dataset?**
     - Training a dataset means using it to build a model. Testing involves evaluating model performance on unseen data.

8. **What is sklearn.preprocessing?**
     - It is a module in Scikit-learn that provides functions for data scaling, normalization, encoding, and transformation.

9. **What is a Test set?**
     - A test set is a portion of data used only for final model evaluation to check how well it generalizes to unseen data.

10. **How do we split data for model fitting (training and testing) in Python?**
     - Using train_test_split() from Scikit-learn.

10.1 **How do you approach a Machine Learning problem?**
    
  - Understand the problem, clean and explore data, choose the right algorithm, train the model, evaluate, and improve.

11. **Why do we have to perform EDA before fitting a model to the data?**
     - Exploratory Data Analysis helps understand data patterns, detect outliers, and decide preprocessing steps.

12. **What is correlation?**
     - It measures the linear relationship between two variables, typically between -1 and +1.

13. **What does negative correlation mean?**
     - It means that one variable increases while the other decreases.

14. **How can you find correlation between variables in Python?**
     - Use df.corr() in pandas to compute correlation between columns.

15. **What is causation? Explain the difference between correlation and causation with an example.**
     - Causation implies one variable causes the other to change.
Example: Ice cream sales and drowning are correlated (due to summer), but one doesn't cause the other.

16. **What is an Optimizer? What are different types of optimizers?**
     - Optimizers adjust model parameters to reduce loss. Examples: Gradient Descent, Adam, RMSprop.

17. **What is sklearn.linear_model?**
     - It's a module in Scikit-learn that provides linear models like Linear Regression, Logistic Regression.

18. **What does model.fit() do? What arguments must be given?**
     - It trains the model. Arguments: training data X and labels y.

19. **What does model.predict() do? What arguments must be given?**
     - It makes predictions on new data. Argument: input data X.

20. **What are continuous and categorical variables?**
     - Continuous: numeric (e.g., age), Categorical: grouped values (e.g., city).

21. **What is feature scaling? How does it help in Machine Learning?**
     - Feature scaling normalizes data to a specific range. It improves model performance and convergence.

22. **How do we perform scaling in Python?**
     - Using StandardScaler or MinMaxScaler from sklearn.preprocessing.

23. **What is sklearn.preprocessing?**
     - It's a library for transforming data: scaling, encoding, imputing missing values.

24. **How do we split data for model fitting (training and testing) in Python?**
     - Using train_test_split(X, y, test_size=0.2) from Scikit-learn.

25. **Explain data encoding.**
     - Data encoding converts categorical data into numerical form. Methods include One-Hot and Label Encoding.
